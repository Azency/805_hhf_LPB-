# 生成离散的$S_t$积分轨道
$S_t = e^{-rt} \cdot max(U_0 e^{gt}, U_0e^{(r-l-\frac{1}{2}\sigma^2)t+\sigma B_t})$

```python
def St_intg_item(N, T_b, T_e, S0, r, l, g, sigma) -> list:
    """
    模拟股票的随机过程
    N : 分区数量
    T_b : 积分下限,以年为单位
    T_e : 积分上限,以年为单位


    S0 : 初始值
    r : 
    l :
    g : 
    sigma :
    """
    dt = (T_e-T_b)/N
    t_point = np.linspace(T_b, T_e, N)

    lgSt = np.zeros(N)
    lgSt[0] = np.log(S0)
    lgSt[1:] = np.cumsum((r - l - 0.5 * sigma**2) * dt + sigma * (np.sqrt(dt)*np.random.standard_normal(size=1)))  # 迭代生成
    St = S0 * np.exp(lgSt)
    # print("st",St)

    final_int = np.zeros(N)
    for i in range(N):
        final_int[i] = np.exp(-r * t_point[i]) * np.maximum(np.exp(g*t_point[i]) - St[i], 0)

    return t_point , final_int
```
